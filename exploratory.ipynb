### Can we predict the length of an order?

```python
orders_prior.head()
```

```python
g = orders_prior.groupby('order_id')
```

```python
n_products = g.count()['product_id']
```

```python
n_products.head()
```

```python
orders['n_products'] = n_products
```

```python
orders.head()
```

```python
orders_prior = orders[orders['eval_set'] == 'prior']
```

```python
orders_prior.shape
```

```python
del orders_prior['eval_set']
```

```python
orders_prior
```

```python
del orders_prior['user_id']
```

```python
orders_prior.head()
```

```python
orders_prior_nonan = orders_prior.dropna()
```

```python
y = orders_prior_nonan['n_products']
```

```python
X = orders_prior_nonan.loc[:, 'order_number': 'days_since_prior_order']
```

```python
X.head()
```

```python
y.head()
```

## Make a model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_jobs=3)
```

```python
rf.fit(X,y)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y, rf.predict(X))
```

```python
y.mean()
```

```python
y.std()
```
